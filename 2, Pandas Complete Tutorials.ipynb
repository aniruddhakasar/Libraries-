# >> Series in Pandas 

```python
list_1 = [1,2,3,-4,-5,"data"]                                           # CREATE LIST 
print(list_1)
a = pd.Series(list_1)                                                   # DISPLAY LIST USING SERIES METHOD 
print(a)
list_2 = [1,2,3,-4,-5,"data"]                                           # CREATE LIST 
print(list_2)
b = pd.Series(list_1, index=["a","b","c","d","e","f"])                  # DISPLAY LIST USING SERIES METHOD WITH INDEX
print(b)
c = pd.Series(list_1, index=["a","b","c","d","e","f"], name= "INFO")    # DISPLAY LIST USING SERIES METHOD WITH INDEX & NAME
print(c)
d = pd.Series({"a":1,"b":2,"c":3})                                      # DISPLAY SERIES USING DICTIONARY  
print(d)
```

# >> DataFrame in Pandas 

```python
a = pd.DataFrame()                                                      # CREATING EMPTY DATAFRAME 
print(a)
b = ["a","b","c","d"]                                                   # CREATING LIST 
print(b)
df = pd.DataFrame(b)                                                    # CREATING DATAFRAME USING LIST 
print(df)                                                               
c = [[1,2,3],[4,5,6],[7,8,9]]                                           # LIST OF LISTS 
df2 = pd.DataFrame(c)                                                   # CREATING DATAFRAME USING NESTED LIST 
print(df2)
d = {"ID":[1,2,3,4],"NO.":[10,20,30,40],"SG":[111,222,333,444]}         # CREATING DICTIONARY
df3 = pd.DataFrame(d)                                                   # CREATING DATAFRAME USING DICTIONARY
print(df3)
```

# >> Methods on Pandas 
###      concat()

```python
sr1 = pd.Series([1,2,3,4])
sr1
```

```python
pd.concat([sr1,sr2])                                        # ADDING 2 SERIES INTO 1 SERIES IN ROW FORM 
```

```python
df14 = pd.DataFrame({"ID":[1,2,3,4], "NAME":["SAI","GAUTI","AJINX","DRAVID"]})     # CREATING 2ND DATAFRAME 
df14 
```

```python
pd.concat([df13,df14], axis = 1)                                     # MERGING 2 DATAFRAME INTO 1 DATAFRAME IN COLUMN FORM 
```

```python
pd.concat([df13,df14], axis = 1 , keys = ['1stDataFrame', '2ndDataFrame'])   # MERGING DATAFRAMES WITH HEADING NAMES 
```

# >> Methods on Pandas 
###      join()

```python
df15 = pd.DataFrame({"ID":[10,20,30,40], "DIV":["A","B","C","D"]})          # CREATING 1ST DATAFRAME 
df15
```

```python
df16 = pd.DataFrame({"ID":[1,2,3,4], "NAME":["SAI","GAUTI","AJINX","DRAVID"]})     # CREATING 2ND DATAFRAME 
df16
```

```python
df16.join(df15)                                             # JOINING 2 DIFFERENT DATAFRAMES INTO 1 DATAFRAME IN COLUMN FORM 
```

```python
df16.join(df15, lsuffix = '_1st')                  # JOINING 2 DTAFRAMES OF SAME COLUMN INTO 1 DATAFRAME IN COLUMN FORM 
```

# >> Methods on Pandas 
###     append()

```python
df17 = pd.DataFrame({"ID":[10,20,30,40], "NAME":["A","B","C","D"]})          # CREATING 1ST DATAFRAME 
df17
df18 = pd.DataFrame({"ID":[1,2,3,4], "NAME":["SAI","GAUTI","AJINX","DRAVID"]})     # CREATING 2ND DATAFRAME 
df18

display(df17,df18)
```

# >> Methods on Pandas 
###      pivot_table()

```python
df19 = pd.DataFrame({"SCORE":[450,437,483,401], "SIXES":[10,8,6,12]})          # CREATING 1ST DATAFRAME 
df19
df20 = pd.DataFrame({"FOURS":[12,16,13,19], "PLAYER":['GAUTI','DRAVID','AJINX','RAHUL']})     # CREATING 2ND DATAFRAME 
df20

display(df19,df20)
```

```python
df21 = pd.concat([df19,df20] , axis = 1)                   # MERGING 2 DATAFRAME INTO 1 DATAFRAME IN COLUMN FORM 
df21                                                                     
```

```python
df21.pivot_table(index = 'PLAYER', columns = 'SCORE' , fill_value = 0)  # USED TO CREATE TABLE VIEW OF DATAFRAME 
```

# >> Methods on Pandas 
###      melt()

```python
pd.melt(df21, id_vars = ['PLAYER'])                                       # USED TO DISPLAY VARIABLES AND VALUES OF DATAFRAME
```

```python
pd.melt(df21, id_vars = ['PLAYER'] , value_vars = 'SCORE')           # USED TO DISPLAY VARIABLES AND VALUES OF SPECIFIC COLUMN
```

```python
pd.melt(df21, id_vars = ['PLAYER'] , value_vars = 'SCORE', var_name = 'TOTAL', value_name = 'RUNS')
                                                                                        # CHANGING NAMES OF VARIABLES AND NAMES
```
